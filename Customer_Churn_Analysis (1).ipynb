```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 

df = pd.read_csv('churn.csv')
print(df.shape)
print(df.head())
print(df.info())
print(df.isnull().sum())
```

```python
churn_rate = df['Churn'].value_counts(normalize=True) * 100
print(churn_rate) 

plt.figure(figsize=(6,6))
df['Churn'].value_counts().plot(kind='pie',
                                autopct='%1.1f%%',
                                labels=['Stayed', 'Churned'],
                                colors=['#66b3ff', '#ff6666'])
plt.title('customer Churn Rate')
plt.ylabel('')
plt.show()
```

```python
sns.boxplot(data=df, x='Churn', y='MonthlyCharges')
plt.title('Monthly Charges vs Churn')
plt.show()
```

```python
print(df.columns.tolist())
```

```python
stayed = df[df['Churn'] == 0]['MonthlyCharges']
churned = df[df['Churn'] == 1]['MonthlyCharges']

plt.figure(figsize=(8,6))
plt.boxplot([stayed, churned], labels=['Stayed', 'Churned'])
plt.title('Monthly Charges vs Churn')
plt.ylabel('Montly Charges')
plt.xlabel('Customer Status')
plt.show()
```

```python
#Chart 1 - Customer Churn Rate (Pie Chart)
plt.figure(figsize=(6,6))
df['Churn'].value_counts().plot(kind='pie',
    autopct='%1.1f%%',
    labels=['Stayed', 'Churned'],
    colors=['#66b3ff', '#ff6666'])
plt.title('Customer Churn Rate')
plt.ylabel('')
plt.show()
```

```python
#Chart 2 - Churn by contract Type 
plt.figure(figsize=(8,6))
sns.countplot(data=df, x='Contract',
              hue='Churn',
              palette=['#66b3ff','#ff6666'])
plt.title('Churn by Contract Type')
plt.xlabel('Contract Type')
plt.ylabel('Number of Customers')
plt.legend(labels=['Stayed', 'Churned'])
plt.show()
```

```python
# Chart 3 - Monthly Charges vs Churn 
stayed = df[df['Churn'] == 0]['MonthlyCharges']
churned = df[df['Churn'] == 1]['MonthlyCharges']

plt.figure(figsize=(8,6))
plt.boxplot([stayed, churned],
            labels=['stayed', 'Churned'])
plt.title('Monthly Charges vs Churn')
plt.ylabel('Monthly Charges ($)')
plt.xlabel('Customer Status')
plt.show()
```

```python
stayed = df[df['Churn'] == 0]['MonthlyCharges'].values
churned = df[df['Churn'] == 1]['MonthlyCharges'].values

fig, ax = plt.subplots(figsize=(8,6)) 
ax.boxplot([stayed, churned])
ax.set_xticklabels(['Stayed', 'Churned'])
ax.set_title('Monthly Charges vs  Churn')
ax.set_ylabel('Monthly Charges ($)')
ax.set_xlabel('Customer status')
plt.show()

print(f"Avengers charges - Stayed: ${stayed.mean():.2f}")
print(f"Average charges - Churned: ${churned.mean():.2f}") 
```

```python
print(df['Churn'].unique())
print(df['Churn'].dtype)
print(df['Churn'].value_counts())
```

```python
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})

print(df['Churn'].unique())
```

```python
stayed = df[df['Churn'] == 0]['MonthlyCharges'].values 
churned = df[df['Churn'] == 1]['MonthlyCharges'].values 

fig, ax = plt.subplots(figsize=(8,6))
ax.boxplot([stayed, churned])
ax.set_xticklabels(['stayed', 'Churned'])
ax.set_title('Monthly Charges vs Churn')
ax.set_ylabel('Monthly Charges ($)')
ax.set_xlabel('Customer Status')
plt.show()

print(f"Average charges - Stayed: ${stayed.mean():.2f}")
print(f"Average charges - Churned: ${churned.mean():.2f}")
```

```python
#Chart 4 - Senior Citizens Churn Rate 
plt.figure(figsize=(8,6))
sns.countplot(data=df, x='SeniorCitizen',
              hue='Churn',
              palette=['#66b3ff','#ff6666'])
plt.title('Senior Citizen Churn Rate')
plt.xlabel('Senior Citizen (0=No, 1=Yes)')
plt.ylabel('Number of Customers')
plt.legend(labels=['Stayed', 'Churned'])
plt.show()
```

```python
#Chart 5 - Services linked to Churn 
services = ['PhoneService', 'InternetService',
            'OnlineSecurity', 'TechSupport']

fig, axes = plt.subplots(2, 2, figsize=(12,8))
fig.suptitle('Services linked to Churn', fontsize=16)

for i, service in enumerate(services):
    sns.countplot(data=df, x=service,
                  hue='Churn',
                  palette=['#66b3ff','#ff6666'],
                  ax=axes[i//2, i%2])
    axes[i//2, i%2].set_title(service)
    axes[i//2, i%2].legend(labels=['Stayed','Churned'])

plt.tight_layout()
plt.show()
```
